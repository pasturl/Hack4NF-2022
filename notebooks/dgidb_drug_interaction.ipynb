```python
import pandas as pd
import httplib2 as http
import json

try:
    from urlparse import urlparse
except ImportError:
    from urllib.parse import urlparse
```

```python
gene = "NF1"
uri = 'https://dgidb.org'
path = f'/api/v2/interactions.json?genes={gene}'

target = urlparse(uri+path)
method = 'GET'
body = ''
headers = {
'Accept': 'application/json',
}

h = http.Http()

response, content = h.request(
target.geturl(),
method,
body,
headers)
data = json.loads(content)
```
