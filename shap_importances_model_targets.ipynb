```python
import pandas as pd
import os
```

```python
models_path = "models/models_1667495052.3974156/"
```

```python
model_targets = os.listdir(models_path)
```

```python
model_targets.pop(0)
```

```python
model_targets[0]
```

```python
df = pd.DataFrame()
for model_target in model_targets:
    df_tmp = pd.read_csv(f"{models_path}{model_target}/feature_importance_shap_lgb.csv", sep=";")
    df_tmp["model_target"] = model_target
    df = pd.concat([df, df_tmp])
```

```python
df.head()
```

```python
df.sort_values("shap_value", ascending=False, inplace=True)
```

```python
df.to_csv(f"{models_path}shap_importances_model_targets.csv", sep=";", index=False)
```
